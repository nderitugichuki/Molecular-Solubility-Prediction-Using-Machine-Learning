MACHINE LERNING PROJECT

```python
#load the libraries
import pandas as pd
```

```python
#load the data
df=pd.read_csv(r'C:\Users\Admin\Desktop\Datasets\d.csv')
df
```

DATA PREPARATION

```python
#Data Separation as X and Y
y = df['logS']
y
```

```python
x = df.drop('logS',axis=1 )
x
```

```python
#Data Splitting
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2,random_state=100)
```

```python
x_test
```

```python
x_train
```

MODEL BUILDING

```python
#Using Linear Regression to train thr model

from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
#Applying the model to make the prediction
y_lr_train_pred=lr.predict(x_train)
y_lr_test_pred=lr.predict(x_test)
```

```python
print(y_lr_train_pred)
```

```python
print(y_lr_test_pred)
```

EVALUATING THE MODEL PERFORMANCE

```python
from sklearn.metrics import mean_squared_error, r2_score
lr_train_mse=mean_squared_error(y_train,y_lr_train_pred)
lr_train_r2=r2_score(y_train,y_lr_train_pred)

lr_test_mse=mean_squared_error(y_test,y_lr_test_pred)
lr_test_r2=r2_score(y_test,y_lr_test_pred)
print('LR MSE (Train):',lr_train_mse)
print('LR R2 (Train):', lr_train_r2)
print('LR MSE (Test):',lr_test_mse)
print('LR R2 (Test):',lr_test_r2)
```

```python
lr_results=pd.DataFrame(['LinearRegression',lr_train_mse, lr_train_r2, lr_test_mse,lr_train_r2]).transpose()
lr_results
```

```python
lr_results=pd.DataFrame(['LinearRegression',lr_train_mse, lr_train_r2, lr_test_mse,lr_train_r2]).transpose()
lr_results.columns=['Method','Training MSE', 'Training R2', 'Test MSE', 'Test R2']
```

```python
lr_results
```

RANDOM FOREST

```python
#Training the model Using Random Forest
from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor(max_depth=2, random_state=100)
rf.fit(x_train,y_train)
```

```python

#Applying the model to make the prediction
y_rf_train_pred=rf.predict(x_train)
y_rf_test_pred=rf.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
rf_train_mse=mean_squared_error(y_train,y_rf_train_pred)
rf_train_r2=r2_score(y_train,y_rf_train_pred)

rf_test_mse=mean_squared_error(y_test,y_rf_test_pred)
rf_test_r2=r2_score(y_test,y_rf_test_pred)
print('LR MSE (Train):',rf_train_mse)
print('LR R2 (Train):', rf_train_r2)
print('LR MSE (Test):',rf_test_mse)
print('LR R2 (Test):',rf_test_r2)
```

```python
rf_results=pd.DataFrame(['Random forest',rf_train_mse, rf_train_r2, rf_test_mse,rf_train_r2]).transpose()
rf_results
```

```python
rf_results=pd.DataFrame(['Random Forest',rf_train_mse, rf_train_r2, rf_test_mse,rf_train_r2]).transpose()
rf_results.columns=['Method','Training MSE', 'Training R2', 'Test MSE', 'Test R2']
```

```python

rf_results
```

MODEL EVALUATION

```python
df_models=pd.concat([lr_results,rf_results],axis=0)
df_models
```
